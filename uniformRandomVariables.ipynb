### Uniform-random-variables

To simulate samples from a uniformm distribution over  [0,1] , a congruent generator is defined from a recursive sequence in the form  ùë•ùëõ+1=(ùëé.ùë•ùëõ+ùëè)%ùëö , where  %  stands for the remainder of integer division. We will consider the case where  ùëé=65539 ,  ùëè=0  and  ùëö=231  that defines the RANDU generator. We intialize the sequence with  ùë•0‚â†0  and thus define iteratively values  ùë•1,ùë•2,‚Ä¶  that are in the set  {1,‚Ä¶,ùëö‚àí1} . Then, the corresponding values  ùë¢ùëõ=ùë•ùëõ/ùëö  are in  [0,1] . We want to check that the samples  ùë¢0,ùë¢1,ùë¢2,‚Ä¶  are approximately distributed according to a uniform distribution over  [0,1] .

1 - Write a function sequence(N) that returns simulated data  ùë¢0,ùë¢1,ùë¢2,‚Ä¶,ùë¢ùëÅ‚àí1 .

2 - Plot the histogram of the data for  ùëÅ=105  (https://en.wikipedia.org/wiki/Histogram). How do you select the number of intervals of the histogram ? Do the data seem to be well described by a uniform distribution ?

Note that RANDU is a poor generator that should not be used in practice ( see https://fr.wikipedia.org/wiki/Mersenne_Twister for a more efficient generator )

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use("fivethirtyeight")

def sequence(N):
    a=65539
    m=2**31
    Un=[]
    Xn=[1]
    for i in range (1,N):
        Xn.append((Xn[i-1]*a)%m)
        Un.append(Xn[i]/m)
    figure(figsize=(8, 6))
    count, bins, ignored = plt.hist(Un,bins="auto",density=True,edgecolor='black', facecolor='g')
    plt.title('Simulation of uniform random variables')
    plt.show()

#I used the arg bins='auto' to define the intervals automatically, 
# other ways i can use (Sturges' formula math.log(N,2)+1) which gives 18 bins versus 48 generated by the arg bins ='auto'
#the data seems to follow a uniform distribution

sequence(10**5)
```
